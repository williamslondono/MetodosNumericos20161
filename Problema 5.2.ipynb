```python
import matplotlib.pyplot as plt 
import math
from tabulate import tabulate
%matplotlib inline

array=[]
tablaB=[]
valores=[]
raices=[]
Error=[]
tabla1=[]
ErrorA=[]
xReal=0.4129
fA, fB=1.0, 1.0
cont=0
xl=0.0 
xu=1.0
xr=0.0
Es=0.1

while(cont<=4):
    fx=5*math.pow(cont,3)-5*math.pow(cont,2)+(6*cont)-2
    valores.append(cont)
    array.append(fx)
    cont+=0.07
    tabla1.append([cont, fx])

def metodo1(x):
    fBis=5*math.pow(x,3)-5*math.pow(x,2)+6*x-2
    return fBis

a=0 
while 1:
    xr=(xl+xu)/2
    raices.append(xr)
    fA=metodo1(xl)
    fB=metodo1(xr)
    erpv=math.fabs(((xReal-xr)/xReal)*100)
    if(a==0):
        ErrorA.append("-----")
        eaprx=100
    else:
        j=a-1
        eaprx=math.fabs(((raices[a]-raices[j])/raices[a])*100)
        ErrorA.append(eaprx)
    tablaB.append([xr, str(erpv)+" %", str(ErrorA[a])+" %"])
    if ((fA*fB)<0):
        xu=xr
    if((fA*fB)>0):
         xl=xr
    a+=1
    if(eaprx<Es):
        break

plt.title("f(x)")
plt.axis([0, 4, -10, 40])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(valores, array, 'm-')
plt.plot(xReal,0, 'rs', label='Raiz Exacta=\n0.4129')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print tabulate(tabla1, headers=["x", "f(x)"], tablefmt="fancy_grid")
print tabulate(tablaB, headers=["Raiz","Error RP", "Error Aprox"], tablefmt="fancy_grid")
```
