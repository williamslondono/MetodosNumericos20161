```python
import math
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2 
%matplotlib inline

n=15
sumaX,sumaYA, sumaYB,sumaProA, sumaProB,cuadrado=0,0,0,0,0,0
tiempo,vMedida=[],[10.0, 16.3, 23.0, 27.5, 31.0, 35.6, 39.0, 41.5, 42.9, 45.0, 46.0, 45.5, 46.0, 49.0, 50.0]
modelo1=[8.953,16.405,22.607,27.769,32.065,35.641,38.627,41.095,43.156,44.872,46.301,47.490,48.479,49.303,49.988]
modelo2=[11.240,18.570,23.729,27.556,30.509,32.855,34.766,36.351,37.687,38.829,39.816,40.678,41.437,42.110,42.712]
valores, ecuacion1, ecuacion2=[],[],[]

for t in range (1, 16):
    tiempo.append(t)
    
def sumProd(aux):
    suma1=0
    for i in range(n):
        suma1+=(vMedida[i]*aux[i])
    return suma1

def sumXY(aux2):
    suma2=0
    suma3=0
    for j in range (n):
        suma2+=vMedida[j]
        suma3+=aux2[j]
    producto=suma2*suma3
    return producto

def raices(aux3):
    raiz1=0
    raiz2=0
    sumx=0
    sumcx=0
    sumy=0
    sumcy=0
    for k in range(n):
        sumx+=math.pow(vMedida[k],2)
        sumcx+=vMedida[k]
        sumy+=math.pow(aux3[k],2)
        sumcy+=aux3[k]
    raiz1=math.sqrt(((n*sumx)-(math.pow(sumcx,2))))
    raiz2=math.sqrt(((n*sumy)-(math.pow(sumcy,2))))
    return (raiz1*raiz2)

def coeficiente(vec):
    r=(((n*sumProd(vec))-sumXY(vec))/(raices(vec)))
    return (r)

print ("El coeficiente de correlacion entre vMedida y el Modelo 1 es: "), coeficiente(modelo1)
print ("El coeficiente de correlacion entre vMedida y el Modelo 2 es: "), coeficiente(modelo2)
print ("Los coeficientes de correlacion (r) son proximos a 1, esto quiere decir, que las relaciones para\nambos casos son fuertes y directas.")

for z in range(n):
    sumaX+=vMedida[z]
    sumaYA+=modelo1[z] 
    sumaYB+=modelo2[z]
    cuadrado+=math.pow(vMedida[z],2)
    sumaProA+=vMedida[z]*modelo1[z]
    sumaProB+=vMedida[z]*modelo2[z]

a1A=((n*sumaProA)-(sumaX*sumaYA))/((n*cuadrado)-(math.pow(sumaX,2)))
a0A=(sumaYA/len(modelo1))-(a1A*(sumaX/len(vMedida)))

a1B=((n*sumaProB)-(sumaX*sumaYB))/((n*cuadrado)-(math.pow(sumaX,2)))
a0B=(sumaYB/len(modelo2))-(a1B*(sumaX/len(vMedida)))

print ("\nPara b) y c) la ecuaci贸n es:\n    Vmodelo=%.3f + %.3fvMedida")%(a0A, a1A)
print ("\nPara b) y c) la ecuaci贸n es:\n    Vmodelo=%.3f + %.3fvMedida")%(a0B, a1B)

for k in range (50):
    valores.append(k)
    f1=a0A+(a1A*k)
    ecuacion1.append(f1)
    f2=a0B+(a1B*k)
    ecuacion2.append(f2)
    
plt.title("Grafica #1")
plt.grid(True)
plt.axis([0,60,0,60])
plt.ylabel('Velocidad Calculada (b)')
plt.xlabel('Velocidad Medida')
plt.plot(vMedida, modelo1, 'co', label="Valores iniciales")
plt.plot(valores, ecuacion1, 'm--', linewidth=2.0, label="y=-0.859+1.032x")
plt.legend(loc = 'lower right', numpoints = 2)
plt.show()

plt2.title("Grafica #2")
plt2.grid(True)
plt2.axis([0,60,0,60])
plt2.ylabel('Velocidad Calculada (c)')
plt2.xlabel('Velocidad Medida')
plt2.plot(vMedida, modelo2, 'ro', label="Valores iniciales")
plt2.plot(valores, ecuacion2, 'b--', linewidth=2.0, label="y=5.776+0.752x")
plt2.legend(loc = 'lower right', numpoints = 2)
plt2.show()
```
