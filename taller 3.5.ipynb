```python
import math 
from tabulate import tabulate


tab = []
ant2=[]
vAprx2 = []
Ev2 = []
Epv2 = []
Ea2 = []
x = 5
n=1
vr = math.pow(math.e, (-x))
value=0.0 
value2=0.0 
valueMet2=0.0 
valMet2 = 0.0
ant = []
vAprx = []
Ev = []
Epv= []
Ea = []
for i in range (0, 20):
    if(i%2==0):
        value = (math.pow(x, i))/math.factorial(i)
        valueAnt = value2
        value2+=value
        ant.append(valueAnt)
        vAprx.append(value2)
    else:
        value = (math.pow(x, i))/math.factorial(i)
        valueAnt = value2
        value2-=value
        ant.append(valueAnt)
        vAprx.append(value2)
    Erv=math.fabs(vr-vAprx[i])
    Ev.append(Erv)
    Erpv=math.fabs((Ev[i]/vr)*100)
    Epv.append(Erpv)
    Eaprx=math.fabs(((vAprx[i]-ant[i])/vAprx[i])*100)
    Ea.append(Eaprx)
   

    valueMet2 = (math.pow(x, i))/math.factorial(i)
    valueAnt2=valMet2
    ant2.append(valueAnt2)
    valMet2+=valueMet2
    valueMet22=n/valMet2
    vAprx2.append(valueMet22)
    Erv2=math.fabs(vr-vAprx2[i])
    Ev2.append(Erv2)
    Erpv2=(Ev2[i]/vr)*100
    Epv2.append(Erpv2)
    Eaprx2=math.fabs(((vAprx2[i]-ant2[i])/vAprx2[i])*100)
    Ea2.append(Eaprx2)  
    tab.append([i+1,vAprx[i],Epv[i],Eaprx, vAprx2[i],Epv2[i], Eaprx2])

                            
print tabulate(tab,headers=["Ter","1Metodo", "ErrPorc", "ErrAprox", "2Metodo", "ErrPorc", "ErrAprox"], tablefmt="fancy_grid")
```
