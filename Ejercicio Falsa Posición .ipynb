```python
import math
from tabulate import tabulate

t=10
v=40
xl=12
xu=16
xr= 0.0
xV=14.7802
g=9.8
m=68.1
tabla=[]
cont=0
fr1=1.0
fr2=1.0
fr3=1.0

def evaluar(num):
    x=((g*m)/num)*(1-math.exp(-(num/68.1)*10.0))-v
    return x

while ((fr1*fr3)!=0):
    fr1=evaluar(xl)
    fr2=evaluar(xu)
    xr=xu-(fr2*(xl-xu))/(fr1-fr2)
    fr3=evaluar(xr)
    erpv=math.fabs(((xV-xr)/xV)*100)
    cont+=1
    tabla.append([cont, xr, str(erpv)+" %"])
    if ((fr1*fr3)<0):
        xu=xr
    if((fr1*fr3)>0):
         xl=xr 

print tabulate(tabla, headers=["Proceso", "Raices", "Error"], tablefmt="fancy_grid")
```
